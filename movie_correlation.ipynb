```python
# Import libraries

import pandas as pd
import seaborn as sns

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read data
df = pd.read_csv(r'C:\Users\Suman\Downloads\movies.csv')
```

```python
df.head()
```

## Data cleaning

```python

df.isnull().sum()
```

```python
df1 = df
df1['budget'] = df['budget'].fillna(df.budget.mode())
```

```python
df1.isnull().sum()
```

```python
df2 = df1.dropna()
```

```python
df2.isnull().sum()
```

```python
df2.shape
```

```python
df.shape
```

```python
# Data types of columns

df2.dtypes
```

```python
# Changing datatype of budget and gross

df2['budget'] = df2['budget'].astype('int64')
df2['gross'] = df2['gross'].astype('int64')
```

```python
df2
```

```python
df2 = df2.sort_values(by=['gross'], inplace=False, ascending=False)
```

## Finding correlations in data

```python
# Hypothesis:
# Budget has high correlation 
# Company has high correlation
```

```python
# scatter plot with budget vs gross

plt.scatter(x=df2['budget'], y=df2['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget of film')
plt.ylabel('Gross Earning')

plt.show()
```

```python
df2.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df2, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Let's start looking at correlation
```

```python
df2.corr(method='pearson') # pearson(default), kendal, spearman
```

```python
 # High correlation between budget and gross
```

```python
correlation_matrix = df2.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix between neumeric features')
plt.xlabel('Movie Features')
plt.xlabel('Movie Features')

plt.show()
```
